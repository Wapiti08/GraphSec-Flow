```python
import matplotlib.pyplot as plt

# Data
dmax = [3, 6, 10]
paths = [0, 3268, 49773104]
runtime = [119346964.2, 72720949.6, 67835533.5]

# ===== Figure 1: Paths vs d_max (log scale) =====
plt.figure(figsize=(6, 4))
plt.plot(dmax, paths, marker="o", linewidth=2)
plt.yscale("log")
plt.xlabel("Maximum Depth (d_max)", fontsize=12)
plt.ylabel("Number of Paths (log scale)", fontsize=12)
plt.title("Impact of d_max on Inferred Propagation Paths", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()

# ===== Figure 2: Runtime vs d_max =====
plt.figure(figsize=(6, 4))
plt.plot(dmax, runtime, marker="o", linewidth=2)
plt.xlabel("Maximum Depth (d_max)", fontsize=12)
plt.ylabel("Runtime (ms)", fontsize=12)
plt.title("Runtime vs Maximum Depth", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
dmax = [3, 6, 10]
paths = [0, 3268, 49773104]
runtime = [119346964.2, 72720949.6, 67835533.5]

# Create a single figure with twin y-axes
fig, ax1 = plt.subplots(figsize=(7, 5))

# --- Left Y-axis: Paths (log scale) ---
ax1.plot(
    dmax, paths,
    marker="o", linewidth=2, color="tab:blue",
    label="Number of Paths"
)
ax1.set_yscale("log")
ax1.set_xlabel("Maximum Depth (d_max)", fontsize=14)
ax1.set_ylabel("Number of Paths (log scale)", color="tab:blue", fontsize=14)
ax1.tick_params(axis='y', labelcolor="tab:blue")
ax1.grid(True, linestyle="--", alpha=0.6)

# --- Right Y-axis: Runtime ---
ax2 = ax1.twinx()
ax2.plot(
    dmax, runtime,
    marker="s", linewidth=2, color="tab:red",
    label="Runtime (ms)"
)
ax2.set_ylabel("Runtime (ms)", color="tab:red", fontsize=12)
ax2.tick_params(axis='y', labelcolor="tab:red")

# Collect legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc="upper left")

plt.title("Impact of Maximum Depth (d_max)", fontsize=14)
plt.tight_layout()
plt.show()
```
