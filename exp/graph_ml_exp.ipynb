```python
from pathlib import Path
import networkx as nx
```

```python
graph_data_path = Path.cwd().parent.joinpath("data", "graph_metric.graphml")
```

```python
G = nx.read_graphml(graph_data_path)
```

```python
# Print the top 10 nodes (you can modify the number as needed)
print("Top 10 nodes:")
top_nodes = list(G.nodes())[:10]
print(top_nodes)
```

```python
# Print the top 10 edges (you can modify the number as needed)
print("Top 10 edges:")
top_edges = list(G.edges())[:10]
print(top_edges)
```
