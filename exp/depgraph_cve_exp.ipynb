```python
import pickle
from pathlib import Path
import networkx as nx
```

```python
dep_graph_path = Path.cwd().parent.joinpath("data", "dep_graph.pkl")
```

```python
with dep_graph_path.open('rb') as fr:
    depgraph = pickle.load(fr)
```

```python
if isinstance(depgraph, nx.DiGraph):
    G = nx.Graph(depgraph.to_undirected())
else:
    G = nx.Graph(depgraph)
print(f"[INFO] Loaded graph: |V|={G.number_of_nodes()}, |E|={G.number_of_edges()}")
```

```python
for nid, attrs in list(G.nodes(data=True))[:10]:
    print(nid, attrs)
```

```python
cve_nodes = [(nid, attrs["CVE"]) for nid, attrs in G.nodes(data=True) if "CVE" in attrs]
print(cve_nodes[:100])
```
